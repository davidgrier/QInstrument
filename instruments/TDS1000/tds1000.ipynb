```python
from QTDS1000 import QTDS1000
import matplotlib.pyplot as plt
```

## Create an instance of the oscilloscope model

```python
scope = QTDS1000().find()
```

## Read a trace from channel 1

```python
t, s = scope.data('CH1')
```

```python
## Plot the trace
```

```python
plt.plot(t, s)
plt.xlabel('time [s]')
plt.ylabel('signal [V]')
```
